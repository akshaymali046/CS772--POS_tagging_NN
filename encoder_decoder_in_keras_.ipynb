```python
from keras.preprocessing.text import Tokenizer
word_tokenizer = Tokenizer()
#not using the encoded form of the words, just using tokenizer to get a mapping
#between words and embeddings as in the next cell
#In short, X_encoded is not used anywhere in training
word_tokenizer.fit_on_texts(word)
```

```python
X_embed = np.zeros((len(words),300))
word_dict = word_tokenizer.word_index
for word,id in word_dict.items():
  try:
    X_embed[id,:] = model[word] #We have the word2vec embeddings of the words in X_embed matrix now
  except KeyError:
    pass #For the words out of vocabulary, the embedding would be [0]*300 array

print(X_embed.shape)
```

```python
tag_tokenizer = Tokenizer()
tag_tokenizer.fit_on_texts(tag)
Y_encoded = tag_tokenizer.texts_to_sequences(tag) #tokenizing Y_tags for one-hot-encoding later
print(tag[0:20], Y_encoded[0:20])
```

```python
Y_list = []
for item in Y_encoded:
  if len(item)!=0:
    Y_list.append(item[0])
  else:
    Y_list.append(0)
print(len(Y_list))
```

```python
from keras.utils.np_utils import to_categorical
Y = to_categorical(Y_list) #one-hot-encoding the Y_tags
Y_arr = np.array(Y)
print(Y_arr.shape)
```

```python
#making the 80:20 split
word_train=X_embed[:int(len(word_tags)*0.8)]
tag_train=Y_arr[:int(len(tag)*0.8)]

word_test=X_embed[int(len(word_tags)*0.8):]
tag_test=Y_arr[int(len(tag)*0.8):]
print(len(word_train),len(tag_train),len(word_test),len(tag_test))
```
